```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ipl = pd.read_csv('IPL Matches 2008-2020.csv')
```

```python
ipl.head()
```

```python
ipl['player_of_match'].value_counts().head()
```

```python
toss_win = ipl['toss_winner'].value_counts()
toss_win
```

```python
#Top-5 teams with Most Toss Wins

plt.figure(figsize=(5,5))
sns.barplot(x=list(ipl['toss_winner'].value_counts())[:5], y=toss_win.keys()[:5], data=ipl)
plt.show()
```

```python
match_win = ipl['winner'].value_counts()
match_win
```

```python
#Top-5 teams with Most Match Wins

sns.barplot(x=list(ipl['winner'].value_counts())[:6], y=match_win.keys()[:6], data=ipl)
plt.show()
```

```python
#Highest Margin win by winning toss and playing first

batting_first = ipl[(ipl['result']=='runs') & (ipl['toss_decision']=='bat')].sort_values(by='result_margin', ascending=False).reset_index()
batting_first[:10]
```

```python
#As we can clearly see Kings XI Punjab is the team which wins by highest margin ever

plt.figure(figsize=(10,5))
plt.grid(True)
sns.barplot(x=batting_first['result_margin'], y=batting_first['toss_winner'], data=batting_first)
plt.show()
```

```python
#Highest Margin win by winning toss and playing second

batting_second = ipl[(ipl['result']=='runs') & (ipl['toss_decision']=='field')].sort_values(by='result_margin', ascending=False)
batting_second[:5]
```

```python
#We can derive from above that MI, RCB & KKR are the top three teams with less marginal difference

plt.figure(figsize=(30,5))
plt.grid(True)
sns.barplot(x=batting_second['toss_winner'], y=batting_second['result_margin'], data=batting_second)
plt.show()
```

```python
ipl2 = pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
```

```python
ipl2.head()
```

```python
runs_per_batsman = ipl2.groupby('batsman').sum().sort_values(by='batsman_runs', ascending=False)[['batsman_runs','extra_runs','total_runs']]
players = (runs_per_batsman.index[:10])
runs = runs_per_batsman['total_runs'][:10]
#runs per batsman in descending
```

```python
runs
```

```python
plt.figure(figsize=(15,4))
plt.title('Highest Runs Makers in Decreasing Order')
sns.barplot(x=players, y=runs)
```

It is clear that Virat Kohli is topping the charts in terms of Highest runs of approx 6000

```python
#Highest Sixes by player

highest_sixes = ipl2[ipl2['batsman_runs']==6].groupby('batsman').count().sort_values(by='batsman_runs', ascending=False)[['batsman_runs']]
highest_sixes.head()
```

```python
#Highest Fours by a player

highest_fours = ipl2[ipl2['batsman_runs']==4].groupby('batsman').count().sort_values(by='batsman_runs', ascending=False)[['batsman_runs']]
highest_fours.head()
```

```python
players6 = (highest_sixes.index[:5])
runs6 = highest_sixes['batsman_runs'][:5]
players4 = (highest_fours.index[:5])
runs4 = highest_fours['batsman_runs'][:5]
```

```python
plt.figure(figsize=(20,4))
plt.subplot(1,2,1)
plt.title("Most No of Sixes")
sns.barplot(x=players6,y=runs6, palette='coolwarm')
plt.subplot(1,2,2)
plt.title("Most No of Fours")
sns.barplot(x=players4,y=runs4, palette='coolwarm')
```

It can be Derived that Virat Kohli is pretty consistent in terms of 4s and 6s as he is in top 5 for most sixes & fours hits
